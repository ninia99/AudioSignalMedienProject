```python
# Creating an array with values from 1 to 20
numbers = [i for i in range(1, 21)]

three_mult = []
two_mult = []
six_mult = []
others = []

for num in numbers:
    if num%2 == 0:
        two_mult.append(num)
    if num%3 == 0:
        three_mult.append(num)
    if num%6 == 0:
        six_mult.append(num)
    else:
        others.append(num)
    # Replacing with %6 + if and else


print(three_mult, two_mult, six_mult, others)
```

```python
import numpy as np

A = [1, 2, 3]

# What is the difference?
print(np.mean(A))
print(sum(A)/len(A))

# Some example operations of numpy

# Creating an array of ones with the exact same shape as A
B = np.ones_like(A)
print(B, B.dtype, B.shape)

# Concatanating two arrays together
C = np.concat((A, B))
print(C)

# Creating a 3x3 matrix filled with random values between 0 and 1
D = np.random.random((3, 3))
print(D)

# Setting all of the values that are equal to 1 in array C to -100.
C[C==1] = -100
print(C)
```

```python
import matplotlib.pyplot as plt
# Defining the time variable between 0 and 10 with 1 step increaments
t = np.arange(start=0, stop=10, step=1)
# Generating random integer values between -2 and 2 with the exact same shape (number of elemnts) as our time variable
F = np.random.randint(low=-2, high=2, size=t.shape)

# Create a line plot
plt.plot(t, F, color='red', marker='o', linestyle='--', label='Random Plot')

# Add title and labels
plt.title("Simple Line Plot")
# X and Y axis labels
plt.xlabel("t (seconds)")
plt.ylabel("F")
# Adding the grid to our plot for easier readibility
plt.grid()
plt.legend();

```
