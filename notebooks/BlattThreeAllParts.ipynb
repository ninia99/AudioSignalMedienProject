```python
import numpy as np

import matplotlib.pylab as plt


def reconstruct ( sample_vals : np.ndarray , fs : int , t_vals : np.ndarray ):
    result = np.zeros_like(t_vals)
    Ns = np.arange(len(sample_vals))
    for i, t in enumerate ( t_vals ):
        reconstructedSample = 0
        for n in range (len(Ns)):
            T = 1 / fs
            reconstructedSample = reconstructedSample + sample_vals[n] * np.sinc((t - n * T) / T)

        # reconstructedSample = np.sum(sample_vals * np.sinc((t - n * (1/fs)) / (1/fs)))

        result[i] = reconstructedSample
    return result

t = np.linspace(0, 3, 10)
t2 = np.linspace(0, 3, 600)

plt.plot(t2, np.sin(t2))
plt.stem(t, np.sin(t))

t3= np.linspace(0, 3, 600)

input_samples = np.zeros_like(t3)
for i, t in enumerate(t3):
    input_samples[i] = np.sin(t)

reconstructed_ = reconstruct( sample_vals = input_samples, fs = len(t3) / 3 , t_vals = t3)
plt.plot ( t3 , reconstructed_ )

```
